```python
data = pd.read_csv("/content/data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
df = df[["Unnamed: 4", "Unnamed: 13", "Unnamed: 14", "Unnamed: 15", "Unnamed: 37"]]
df.head()
```

```python
df.columns = ["District", "Producer", "Brand", "Salt Type", "Average ppm"]
df.head()
```

```python
df["District"] = df["District"].fillna("Unknown")
df["Producer"] = df["Producer"].fillna("Unknown")
df["Brand"] = df["Brand"].fillna("Unknown")
df["Salt Type"] = df["Salt Type"].fillna("Unknown")
df["Average ppm"] = df["Average ppm"].fillna("0")
```

```python
df = df.iloc[3:]
df.head(10)
```

```python
df["Average ppm"] = df["Average ppm"].astype(float)
df["Average ppm"] = df["Average ppm"].round(1)
```

```python
import numpy as np

def get_top_best_producers(df, min_ppm, max_ppm):
    filtered_df = df[(df["Average ppm"] >= min_ppm) & (df["Average ppm"] <= max_ppm)]
    median_ppm = filtered_df["Average ppm"].median()
    filtered_df["Distance to Median"] = np.abs(filtered_df["Average ppm"] - median_ppm)
    top_producers = filtered_df.sort_values(by="Distance to Median").head(10)
    return top_producers

# Get top 10 best producers (15-30 ppm)
top_best_producers = get_top_best_producers(df, 15, 30)

# Get top 10 worst producers (<15 ppm) excluding zero values
top_worst_producers = df[(df["Average ppm"] < 15) & (df["Average ppm"] > 0)].sort_values(by="Average ppm", ascending=True).head(10)

# Get top 10 over performing producers (>30 ppm)
top_overperforming_producers = df[df["Average ppm"] > 30].sort_values(by="Average ppm", ascending=False).head(10)

print("Top 10 Best Producers (15-30 ppm):")
print(top_best_producers)

print("\nTop 10 Worst Producers (<15 ppm):")
print(top_worst_producers)

print("\nTop 10 Overperforming Producers (>30 ppm):")
print(top_overperforming_producers)
```

```python
producer_performance = df.groupby('Producer')['Average ppm'].agg(
    less_than_15=lambda x: (x < 15).sum(),
    between_15_30=lambda x: ((x >= 15) & (x <= 30)).sum(),
    greater_than_30=lambda x: (x > 30).sum()
).reset_index()

# Calculate total number of samples for each producer
producer_performance['Total_samples'] = producer_performance[['less_than_15', 'between_15_30', 'greater_than_30']].sum(axis=1)

# Calculate the percentage of samples with average ppm between 15-30
producer_performance['Percentage_15_30'] = (producer_performance['between_15_30'] / producer_performance['Total_samples']) * 100

# Reorder columns
producer_performance = producer_performance[['Producer', 'Total_samples', 'less_than_15', 'between_15_30', 'greater_than_30', 'Percentage_15_30']]

# Save results to an Excel file
producer_performance.to_excel('producer_performance.xlsx', index=False)

# Display the results
print(producer_performance)
```

```python
brand_performance = df.groupby('Brand')['Average ppm'].agg(
    less_than_15 = lambda x: (x < 15).sum(),
    between_15_30 = lambda x: ((x >= 15) & (x <= 30)).sum(),
    greater_than_30 = lambda x: (x > 30).sum()
).reset_index()

print("Brand wise performance:")
print(brand_performance)

brand_performance.to_excel('brand_performance.xlsx', index=False)
```

```python
brand_district_performance = df.groupby(['District', 'Brand'])['Average ppm'].agg(
    less_than_15=lambda x: (x < 15).sum(),
    between_15_30=lambda x: ((x >= 15) & (x <= 30)).sum(),
    greater_than_30=lambda x: (x > 30).sum()
).reset_index()

# Function to find best, worst, and overperforming brands
def find_best_worst_overperforming(group):
    best_brand = group.loc[(group['between_15_30'] > 0)].sort_values(by='between_15_30', ascending=False).head(1)
    worst_brand = group.loc[(group['less_than_15'] > 0)].sort_values(by='less_than_15', ascending=False).head(1)
    overperforming_brand = group.loc[(group['greater_than_30'] > 0)].sort_values(by='greater_than_30', ascending=False).head(1)
    return pd.concat([best_brand, worst_brand, overperforming_brand])

# Apply the function to each district
best_worst_overperforming_brands = brand_district_performance.groupby('District').apply(find_best_worst_overperforming).reset_index(drop=True)

# Print the results
print("Best, Worst, and Overperforming Brands Across Each District:")
print(best_worst_overperforming_brands)

# Save to Excel
best_worst_overperforming_brands.to_excel('best_worst_overperforming_brands.xlsx', index=False)
```

```python
brand_district_performance = df.groupby(['District', 'Brand'])['Average ppm'].agg(
    less_than_15=lambda x: (x < 15).sum(),
    between_15_30=lambda x: ((x >= 15) & (x <= 30)).sum(),
    greater_than_30=lambda x: (x > 30).sum()
).reset_index()

# Function to find best, worst, and overperforming brands
def find_best_worst_overperforming(group):
    best_brand = group.loc[(group['between_15_30'] > 0)].sort_values(by='between_15_30', ascending=False).head(1)
    worst_brand = group.loc[(group['less_than_15'] > 0)].sort_values(by='less_than_15', ascending=False).head(1)
    overperforming_brand = group.loc[(group['greater_than_30'] > 0)].sort_values(by='greater_than_30', ascending=False).head(1)
    return best_brand, worst_brand, overperforming_brand

# Initialize lists to store results
districts = []
best_brands = []
worst_brands = []
overperforming_brands = []

# Apply the function to each district
for district, group in brand_district_performance.groupby('District'):
    best_brand, worst_brand, overperforming_brand = find_best_worst_overperforming(group)
    districts.append(district)
    best_brands.append(best_brand['Brand'].values[0] if not best_brand.empty else "N/A")
    worst_brands.append(worst_brand['Brand'].values[0] if not worst_brand.empty else "N/A")
    overperforming_brands.append(overperforming_brand['Brand'].values[0] if not overperforming_brand.empty else "N/A")

# Create the resulting DataFrame
result_df = pd.DataFrame({
    'District': districts,
    'Best Performing Brand': best_brands,
    'Worst Performing Brand': worst_brands,
    'Overperforming Brand': overperforming_brands
})

# Print the results
print("Best, Worst, and Overperforming Brands Across Each District:")
print(result_df)

# Save to Excel
result_df.to_excel('best_worst_overperforming_brands_summary.xlsx', index=False)
```

```python
aggregated_data = df.groupby(['District', 'Producer']).agg(
    total_samples=('Average ppm', 'count'),
    samples_below_15=('Average ppm', lambda x: (x < 15).sum()),
    samples_15_to_30=('Average ppm', lambda x: ((x >= 15) & (x <= 30)).sum()),
    samples_above_30=('Average ppm', lambda x: (x > 30).sum())
).reset_index()

# Determine top 10 producers based on different criteria
top_best_producers = aggregated_data.sort_values(by='samples_15_to_30', ascending=False).head(10)
top_worst_producers = aggregated_data.sort_values(by='samples_below_15', ascending=False).head(10)
top_overperforming_producers = aggregated_data.sort_values(by='samples_above_30', ascending=False).head(10)

# Save results to an Excel file
with pd.ExcelWriter('top_producers_analysis.xlsx') as writer:
    top_best_producers.to_excel(writer, sheet_name='Top 10 Best Producers', index=False)
    top_worst_producers.to_excel(writer, sheet_name='Top 10 Worst Producers', index=False)
    top_overperforming_producers.to_excel(writer, sheet_name='Top 10 Overperforming Producers', index=False)

# Display the top producers for each category
print("Top 10 Best Producers (15-30 ppm):")
print(top_best_producers)
print("\nTop 10 Worst Producers (<15 ppm):")
print(top_worst_producers)
print("\nTop 10 Overperforming Producers (>30 ppm):")
print(top_overperforming_producers)
```
